## «Кирпичики» реляционной БД

Основные понятия, с которыми вы познакомитесь при работе с БД, — таблицы, записи, объекты, атрибуты, ключи, сущности, отношения.

<b>Таблицу БД</b> можно представить как набор однородных записей, т. е. записей об объектах одного типа и с одинаковым набором атрибутов.

<b>Запись</b> — это совокупность нескольких разнородных полей, описывающая некоторую сущность (объект) предметной области. Например, запись «сотрудник» содержит поля id, Name, City, Age и Education.


<b>Объект</b> — это одна из логических единиц хранения информации в базе данных. Используется для сущностей, которые присутствуют в реальной жизни как некоторый предмет, существо или понятие, а не как совокупность каких-либо значений. Например, студент, сотрудник или организация.

<b>Атрибут</b> содержит описание свойств объекта. Например, объект «сотрудник» характеризуется атрибутами «имя», «город проживания», «возраст», «образование» и др.

## Индексирование

При работе с таблицами в реляционных базах данных необходимо, чтобы каждая таблица имела так называемый первичный ключ.

<b>Первичный ключ</b> — это поле, которое используется для обеспечения уникальности данных в таблице. Это значит, что значение в поле первичного ключа в каждой строке (записи) таблицы должно быть уникальным.

В реляционной модели данных таблицы могут быть связаны между собой. Такие связи называются отношениями. Например, для таблиц «Работник» и «Зарплата» можно установить связь по полю «Табельный номер».

При реализации связи между таблицами возникает понятие внешнего ключа. 

<b>Внешний ключ (вторичный)</b> — это столбец (или группа столбцов), используемый в реляционной БД для связи данных между таблицами.

Внешний ключ таблицы соответствует значению первичного ключа таблицы, с которой он связан. Это позволяет сохранять согласованность базы данных путем обеспечения так называемой ссылочной целостности (referential integrity).

Первичный и внешний ключи могут быть как простыми или составными (сложными):

-    Простые ключи — это ключи, которые содержат только одно поле (один атрибут).
-    Составные (сложные) ключи — это ключи, которые содержат несколько полей (атрибутов).

Связи между таблицами позволяют обеспечить целостность данных, чтобы в БД не было потерянных записей.

## Типы отношений в таблице

Выделяют следующие типы отношений:

-    Многие ко многим. Инвестор может вложиться в акции нескольких компаний (многих). Инвесторами какой-то компании являются определенные люди (многие).
-    Один ко многим. У одного клиента может быть несколько номеров.
-    Один к одному. У одного гражданина может быть только один номер паспорта и его номер для каждого гражданина уникален.
