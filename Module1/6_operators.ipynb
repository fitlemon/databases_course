Цель занятия	
- Рассмотреть типы операторов в SQL, научиться работать с операторами описания данных и освоить работу с операторами манипуляции данными.

#### План занятия	
- Типы операторов SQL
- Операторы описания данных
- Операторы манипуляции данными
- Операторы задания прав доступа в БД
- Операторы защиты, восстановления данных.

В SQL существует разграничение операторов по типам в зависимости от выполняемых функций. Выделяют следующие типы: операторы описания данных (DDL, Data Definition Language), операторы манипуляции данными (DML, Data Manipulation Language), операторы задания прав доступа (DCL, Data Control Language) и операторы защиты, восстановления данных (TCL, Transaction Control Language).

Операторы описания данных предназначены для описания (создания), изменения описания и уничтожения объектов базы данных.

- CREATE — для создания нового объекта (например, таблицы) в существующей базе данных.
- ALTER — изменение уже существующего объекта.
- DROP — удаление объекта из базы.

#### Примеры:


```
CREATE TABLE tableName (
    column1 datatype,
    column2 datatype,
    column3 datatype,
   ....
);

ALTER TABLE tableName
ADD columnName datatype;

DROP TABLE tableName;

К операциям манипулирования данными относятся операция добавления или ввода новых записей, операция удаления записей и операция обновления записей. В эту группу входят следующие операторы:

- SELECT — отобрать строки из таблиц;
- INSERT — добавить строки в таблицу;
- UPDATE — изменить строки в таблице;
- DELETE — удалить строки в таблице.

Операторы задания прав доступа предназначены для определения доступа к данным.

- GRANT — предоставляет доступ к объекту.
- REVOKE — отзывает выданное ранее разрешение на доступ.
- DENY — запрет, который прекращает действие разрешения.

Операторы защиты, восстановления данных (TCL, Transaction Control Language) используются для управления транзакциями. Под транзакцией подразумевается команда или несколько команд (инструкций), которые успешно завершаются как единое целое.
- BEGIN TRANSACTION — служит для определения начала транзакции;
- COMMIT TRANSACTION — применяет транзакцию;
- ROLLBACK TRANSACTION — откатывает все изменения, сделанные в контексте текущей транзакции;
- SAVE TRANSACTION — устанавливает промежуточную точку сохранения внутри транзакции.

Итоги занятия:
- Операторы SQL можно разделить на 4 группы: операторы описания данных, операторы манипуляции данными, операторы задания прав доступа в базе данных и операторы защиты, восстановления данных.
- Операторы описания данных служат для создания, изменения и удаления объектов.
- Операторы манипуляции данными предназначены для выборки данных и изменения записей в таблицах БД.
- Операторы задания прав доступа используются для определения прав пользователей относительно объектов базы данных.
- Операторы защиты, восстановления данных позволяют управлять транзакциями.
