## СУБД

СУБД — комплекс программно-языковых средств, позволяющих создать базы данных и управлять данными.

<div style="background-color: #e0fff3; padding: 15px; color: black; width: 80%;">  <b>БД</b> — это история про хранение данных, а СУБД — про управление данными.</div>

При работе с БД СУБД выполняет следующие функции:

-    Поддержка языков программирования для БД.
-    Управление объемами данных вне оперативной памяти.
-    Управление данными в оперативной памяти с использованием дискового кэша.
-    Управление операциями над базой данных (транзакциями).
-    Хранение данных и операций над ними (журнализация).
-    Администрирование БД.

Рассмотрим примеры СУБД.

<b>PostgreSQL</b> — популярная и бесплатная система СУБД для реляционных баз данных. Наибольшее применение нашла для управления БД веб-сайтов и различных сервисов. PostgreSQL представляет собой объектно-реляционную СУБД, что дает ей некоторые преимущества над другими бесплатными СУБД, в большинстве являющимися реляционными.

<b>MySQL</b> — реляционная СУБД с открытым исходным кодом, главными плюсами которой являются ее скорость и гибкость, которая обеспечена поддержкой большого количества различных типов таблиц. MySQL — бесплатная система с простым интерфейсом и возможностью синхронизации с другими базами данных. Это позволяют использовать MySQL как крупным корпорациям, так и небольшим компаниям.

<b>Oracle Database</b> — это объектно-реляционная система управления базами данных (СУБД) от компании Oracle. Первая версия этой объектно-реляционной СУБД появилась в конце 70-х гг., и с тех пор зарекомендовала себя как надежная, функциональная и практичная.

СУБД Oracle постоянно развивается и дорабатывается, упрощая установку и первоначальную настройку и расширяя функционал. Заметный минус Oracle — высокая стоимость лицензии. Поэтому она используется в основном крупными компаниями и корпорациями, которые работают с огромными объемами данных.

<b>SQLite</b> — это встроенная библиотека, которая реализует автономный, безсерверный, нулевой конфигурации, транзакционный механизм СУБД SQL. Нулевая конфигурация означает, что не нужно выполнять дополнительные настройки этой СУБД в вашей системе.

## SQL

Для работы с данными используется язык SQL.

<div style="background-color: #e0fff3; padding: 15px; color: black; width: 80%;">  <b>SQL ([,es kju:' el] англ. structured query language</b> — «язык структурированных запросов») — это декларативный язык программирования, применяемый для создания, модификации и управления данными в реляционной базе данных, управляемой соответствующей СУБД.</div>

SQL используется для:

-    создания в базе данных новой таблицы;
-    добавления в таблицу новых записей;
-    изменения записей;
-    удаления записей;
-    выборки записей из одной или нескольких таблиц по заданному условию;
-    изменения структур таблиц — например, изменение типа поля (столбца) или добавления нового поля;
-    удаления таблиц или других объектов БД.


Итоги занятия:

-    СУБД используется для управления данными в БД — создание, изменение, удаление данных.
-    Среди СУБД для реляционных БД наиболее популярны SQLite, PostgreSQL, MySQL и Microsoft SQL Server.
-    SQL (сокращение от англ. Structured Query Language) — это язык запросов, который применяют, чтобы работать с базами данных, структурированных особым образом.

Далее на курсе будем работать с PostgreSQL.
